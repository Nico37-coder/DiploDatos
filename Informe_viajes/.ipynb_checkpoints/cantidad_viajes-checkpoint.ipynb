```python
B=subset_dl
B.head()
B['Fecha'] = pd.to_datetime(B['Fecha'], format="%m/%y")
B.index = B['Fecha']
df_B = B.resample('M').count()
df_B.head()
data_B= pd.DataFrame(df_B["Mes"])
data_B.head()
#data_B.shape
```

```python
data_A.columns=["Corta distancia"]
data_B.columns=["Larga distancia"]
data_new.columns=["Total"]
```

```python
data_new.head()
```

```python
df_innerAB = data_A.merge(data_B, how='inner', left_index=True, right_index=True)
```

```python
df_innerAB.head()
```

```python
df_innerAB['index'] = df_innerAB.index #Cambiar nombre de columna Index a fecha!!!!
df_innerAB['index'] = pd.to_datetime(df_innerAB['index'])
df_innerAB["Total"] = df_innerAB["Corta distancia"] + df_innerAB["Larga distancia"]
df_innerAB["%Cd"] = (df_innerAB["Corta distancia"] / df_innerAB["Total"]) * 100
df_innerAB["%Ld"] = (df_innerAB["Larga distancia"] / df_innerAB["Total"]) * 100
df_innerAB["%Total"] = (df_innerAB["Total"] / df_innerAB["Total"]) * 100
df_innerAB.head()
```

```python
df_innerAB = df_innerAB[["index","Corta distancia","%Cd","Larga distancia","%Ld","Total","%Total"]]
df_innerAB.sort_values(by='index',ascending=False)  #transformar en fecha datatime index y cambiar vel nombre
#df_innerAB.columns= df_innerAB[["Fecha","Corta distancia","%Cd","Larga distancia","%Ld","Total","%Total"]]
df_innerAB.tail()
```

```python
df_innerAB["Media_To"] = np.mean(df_innerAB["Total"])
df_innerAB["Media_Cd"] = np.mean(df_innerAB["Corta distancia"])
df_innerAB["Media_Ld"] = np.mean(df_innerAB["Larga distancia"])
df_innerAB.head()
```
