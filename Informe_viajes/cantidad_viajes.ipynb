```python
# Separar columna Fecha en Mes y A√±o.
data["Fecha"] = pd.to_datetime(data["Fecha"]) 
data["A√±o"] = data["Fecha"].dt.year 
data["Mes"] = data["Fecha"].dt.month 
data.head()
```

# Armado de informe viajes corta y larga distancia

**Obejtivo del informe:**
Contabilizar en forma mensual los viajes de glucosa del cliente ARCOR. Clasificandolos en viajes de corta y larga distancia.

```python
# Filtrar por cliente "ARCOR S.A.I.C." e "INGRECOR S.A.".
df_mask = data[(data["Cliente"] == "ARCOR S.A.I.C.") | (data["Cliente"] == "INGRECOR S.A.")] 
data = df_mask
data.head()
```

```python
# Filtrar por Origen igual "ARROYITO-ARCOR" y Mercaderia igual "GLUCOSA".
df_mask = data[(data["Origen"] == "ARROYITO-ARCOR") & (data["Mercaderia"] == "GLUCOSA")]
data = df_mask
data.head()
```

```python
data.shape
```

```python
# Filtrar viajes sin estadias y ajustes.
data = data[~data.NroViaje.str.contains("Ajus|EST", na=False, case=False)]
data.shape
```

```python
# Eliminar a√±o 2015 y Mercaderia igual a "VACIOS".
data = data.drop(data[data["A√±o"]==2015].index) 
data = data.drop(data[data["Mercaderia"]=="VACIOS"].index) 
```

```python
data.head()
```

```python
data.shape
```

```python
# Verificar que no existan filas en la columna Mercaderia con "VACIOS".
vacios = (data[data["Mercaderia"]=="VACIOS"].index)
df_vacios = pd.DataFrame(vacios)
len(df_vacios)
```

```python
# Crear un subconjunto con viajes de corta distancia(considerados los menores o iguales a 150 kilometros).
df_mask = data["Km"]<=150 
subset_ds = data[df_mask]
```

```python
len(subset_ds)
```

```python
# Crear un subconjunto con viajes de larga  distancia(considerados los mayores a 150 kilometros).
df_mask = data["Km"]>150
subset_dl = data[df_mask]
```

```python
len(subset_dl)
```

```python
# Control de suma de viajes subconjuntos vs conjunto. Prueba OK.
len(subset_ds) + len(subset_dl) - len(data) 
```

```python
# Renombrar subset_ds con A para simplificar.
A = subset_ds
# Transformar columna Fecha en index del dataframe.
A["Fecha"] = pd.to_datetime(A["Fecha"], format="%m/%y")
A.index = A["Fecha"]
# Aplicar metodo resample para agrupar y contar por mes la cantidad de los viajes.
A = A.resample("M").count()
A = pd.DataFrame(A["Mes"])
A.head()
```

```python
# Renombrar subset_dl con B para simplificar.
B = subset_dl
B.head()
# Transformar columna Fecha en index del dataframe.
B["Fecha"] = pd.to_datetime(B["Fecha"], format="%m/%y")
B.index = B["Fecha"]
# Aplicar metodo resample para agrupar y contar por mes la cantidad de los viajes.
B = B.resample("M").count()
B = pd.DataFrame(B["Mes"])
B.head()
```

```python
A.columns = ["Corta distancia"]
B.columns = ["Larga distancia"]
```

```python
# Unir en un solo dataframe los viajes de corta(A) y larga distancia(B).
df_innerAB = A.merge(B, how="inner", left_index=True, right_index=True)
```

```python
df_innerAB.head()
```

```python
df_innerAB["index"] = df_innerAB.index #Cambiar nombre de columna Index a fecha!!!!
df_innerAB["index"] = pd.to_datetime(df_innerAB["index"])
df_innerAB["Total"] = df_innerAB["Corta distancia"] + df_innerAB["Larga distancia"]
df_innerAB["%Cd"] = (df_innerAB["Corta distancia"] / df_innerAB["Total"]) * 100
df_innerAB["%Ld"] = (df_innerAB["Larga distancia"] / df_innerAB["Total"]) * 100
df_innerAB["%Total"] = (df_innerAB["Total"] / df_innerAB["Total"]) * 100
df_innerAB.head()
```

```python
# Ordenar las columnas.
df_innerAB = df_innerAB[["index","Corta distancia",
                         "Larga distancia","Total",
                         "%Cd","%Ld","%Total"
                         ]]
df_innerAB.sort_values(by="index",ascending=False)  
df_innerAB.tail()
```

```python
# Calcular la media de los viajes. 
df_innerAB["Media_Cd"] = np.mean(df_innerAB["Corta distancia"])
df_innerAB["Media_Ld"] = np.mean(df_innerAB["Larga distancia"])
df_innerAB["Media_To"] = np.mean(df_innerAB["Total"])
df_innerAB.head()
```
